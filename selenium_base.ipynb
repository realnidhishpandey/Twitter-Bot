```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.support.ui import Select
import time
website = 'https://www.adamchoi.co.uk/overs/detailed'
path = 'C:/Users/croma/OneDrive/Desktop/VS code/chromedriver_win32/chromedriver.exe'
driver = webdriver.Chrome(path)
# opens the website
driver.get(website)

# clicks on the all matches button
all_match_button = driver.find_element_by_xpath('//label[@analytics-event = "All matches"]')
all_match_button.click()



# clicks on home away button
# home_away_button = driver.find_element_by_xpath('//label[@analytics-event = "Home/Away split"]')
# home_away_button.click()

# for drop down import from selenium.webdriver.support.ui import Select
drop_down = Select(driver.find_element_by_id('country'))
drop_down.select_by_visible_text('Italy')

time.sleep(3)

matches = driver.find_elements_by_tag_name('tr')

date = []
home_team = []
score_sheet = []
away_team = []

for match in matches:
    
    days = match.find_element_by_xpath('./td[1]').text
    date.append(days)
    # print(days)
    home = match.find_element_by_xpath('./td[2]').text
    home_team.append(home)
    print(home)
    score = match.find_element_by_xpath('./td[3]').text
    score_sheet.append(score)
    # print(score)
    away = match.find_element_by_xpath('./td[4]').text
    away_team.append(away)
    # print(away)

driver.quit()

df = pd.DataFrame({'date': date , 'home_team': home_team , 'score_sheet': score_sheet , 'away_team' : away_team})
df.to_csv('football_data.csv' , index = False)
print(df)
```
